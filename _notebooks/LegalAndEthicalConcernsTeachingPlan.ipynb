# Lesson 1: Key Vocabulary

Before we begin, review these terms. Use our flashcards to study them.

1. **AI Art**: Art created with AI algorithms and machine learning.
2. **Creative Commons**: Licensing system to share work under specific permissions.
3. **Digital Divide**: The gap between those with and without access to digital tech.
4. **Echo Chambers**: Online spaces reinforcing existing viewpoints.
5. **Fair Use**: Doctrine allowing limited use of copyrighted material.
6. **First Amendment**: U.S. constitutional guarantees of free speech, press, etc.
7. **Intellectual Property**: Creations of the mind protected by law.
8. **Misinformation**: False or inaccurate information spread intentionally or not.
9. **Moral Rights**: Creator rights to be credited and protect integrity of work.
10. **Open Sourcing**: Making source code publicly viewable and modifiable.
11. **Open Access**: Unrestricted online access to scholarly works.
12. **Plagiarism**: Using someone else’s work without proper credit.

**Hack:** Use flashcards to memorize all key terms and definitions.

# Lesson 2: Intellectual Property

> Defined by WIPO as “creations of the mind.”

**Examples:** inventions, artistic works, designs, symbols, names.

- Materials created on a computer belong to the creator unless hired work.
- Challenges: easy online access leads to unauthorized use.

**Protections:**
- **Digital Rights Management (DRM)**
- **Watermarking**
- **Licensing Agreements**

**Types of IP Rights:**
- **Copyright**: protects artistic works.
- **Patents**: protects inventions for ~20 years.
- **Trademarks**: protects symbols and names.
- **Trade Secrets**: protects confidential business info.

**Hack:** List ways to protect intellectual property from the lecture.

# Lesson 3: Plagiarism & MIT License

**Plagiarism:** unauthorized use of others’ code or ideas without attribution.

**Forms:**
- Copying code verbatim.
- Minor modifications of others’ code.
- Using AI-generated or purchased code without credit.
- Unauthorized collaboration.

**Consequences:** legal (copyright infringement), academic penalties.

**AP CSP Policy:** unattributed code → zero score.

**MIT License:**
```text
Copyright (c) [Year] [Author]
Permission is granted free of charge to use, copy, modify...
THE SOFTWARE IS PROVIDED “AS IS”, WITHOUT WARRANTY...
```

**Hack:** Define an MIT License for your own repo.

# Lesson 4: Legal Ways to Use IP of Others

- **Creative Commons (CC):** public copyright licenses with six levels.
- **Fair Use:** limited use for education, news, commentary, parody.
- **Open Sourcing vs Open Access:** software vs research articles.

# Lesson 5: Copyright & Creative Commons

**Copyright Rights:** economic (financial) and moral (authorship).

**Prevention Tips:**
- Check licenses.
- Cite sources.
- Use public domain or CC content.
- Seek permission.

**Creative Commons Licenses:**
- CC BY, CC BY-SA, CC BY-ND, CC BY-NC, CC BY-NC-SA, CC BY-NC-ND.

**Hack:** With a partner, write ways to prevent copyright infringement and your partner’s name.
